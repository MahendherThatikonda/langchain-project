```python
from langchain_community.document_loaders import PyPDFLoader
loader=PyPDFLoader("Your booking is confirmed - #334392516.pdf")
docs=loader.load()
docs
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000,chunk_overlap=20)
documents = text_splitter.split_documents(docs)
documents[:5]
```

```python
# Vector Embedding and Vector Store
from langchain_community.embeddings import OpenAIEmbeddings
from langchain_community.embeddings import OllamaEmbeddings
from langchain_community.vectorstores import FAISS
db=FAISS.from_documents(documents[:20],OpenAIEmbeddings())
db
```

```python
## Vector Database
query="By what time the kayaking starts?"
result=db.similarity_search(query)
result[0].page_content
```

```python
#from langchain_community.llms import Ollama
#from langchain_ollama import OllamaLLM
from langchain_openai import ChatOpenAI
#LLM Ollama
llm = ChatOpenAI(model="gpt-4o-mini")
llm
```

```python
# Develop ChatPrompt Template
from langchain_core.prompts import ChatPromptTemplate
prompt = ChatPromptTemplate.from_template(
    """Answer the following question based only on the provided context.
    Think step by step before providing a detailed answer.
    I'll tip you 1000$ if the user finds the answer helpful.
    <context>
    {context}
    </context>
    Question:{input}
"""
)
```

```python
## Chain Introduction
## Create Stuff Document Chain
#%pip install langchain_classic
#from langchain  import rate_limiters
from langchain_classic.chains.combine_documents import create_stuff_documents_chain

document_chain = create_stuff_documents_chain(llm,prompt)
#from langchain.chains import create_retrieval_chain
 #.chains.combine_documents import create_stuff_documents_chain

```

```python
retriever=db.as_retriever()
retriever
```

```python
from langchain_classic.chains import create_retrieval_chain
retrieval_chain = create_retrieval_chain(retriever,document_chain)
```

```python
response=retrieval_chain.invoke({"input":"By what time the kayaking starts?"})
response['answer']
```
